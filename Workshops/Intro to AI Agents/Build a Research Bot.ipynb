```python
# We'll make this a function for easy access
# Use response.choices[0].message.content to filter only response

def call_LLM(user_prompt):
  client = openai.OpenAI(
      api_key = userdata.get("KEY"),
      base_url="https://api.ai.it.ufl.edu"
  )

  response = client.chat.completions.create(
      model="gpt-oss-20b",
      messages = [
          {
              "content": user_prompt
          }
      ]
  )
  return response.choices[0].message.content
```

```python
# Google Drive API
drive_service = build('drive', 'v3')

# Google Slides API
slides_service = build('slides', 'v1')
```

```python
def fetch_wikipedia_summary(topic):

    print(f"üîç Fetching Wikipedia summary for: {topic}")

    wiki = wikipediaapi.Wikipedia('MyAIagent/1.0 (colab@example.com)')
    page = wiki.page(topic)

    if not page.exists():
        return "No Wikipedia page found."

    print("‚úÖ Wikipedia page found.")

    return page.summary[:10000]
```

```python
# Test Wikipedia Function
fetch_wikipedia_summary("Lebron James")
```

```python
def generate_slide_text(topic, wiki_text):
    slides_prompt = f"""
    You are an AI assistant creating a 8-slide Google Slides presentation on "{topic}".
    Use the Wikipedia text below to generate content.

    Format your response exactly like this (no JSON, no markdown):
    Slide 1: <title>
    - Bullet 1
    - Bullet 2
    - Bullet 3
    Slide 2: <title>
    - Bullet 1
    - Bullet 2
    - Bullet 3
    ...
    Wikipedia text:
    {wiki_text}
    """

    text = call_LLM(slides_prompt)
    print("‚úÖ Slides generated successfully.")
    return text
```

```python
# Test generate_slide_text function
generate_slide_text("Lebron James", fetch_wikipedia_summary("Lebron James"))
```

```python
def create_google_slides_with_theme(presentation_title, slide_text):
    print("üßæ Creating Slides")

    template_id = "10AgR2x2mDzqwEXmGplBbSMjTVKo_fRa0PU695yq6Z7s" # Can customize with your choice of template slideshow
    presentation_id = drive_service.files().copy(
        fileId=template_id, body={"name": presentation_title}
    ).execute()["id"]

    # Parse LLM output -> slide data
    slides_data = []
    for section in slide_text.split("Slide ")[1:]:
        lines = section.strip().splitlines()
        if not lines:
            continue
        title = lines[0].split(":", 1)[-1].strip()
        bullets = [l.strip("- ").strip() for l in lines[1:] if l.startswith("-")]
        slides_data.append((title, bullets))

    # Ensure exactly 8 slides always
    while len(slides_data) < 8:
        slides_data.append((" ", []))
    slides_data = slides_data[:8]

    # Create 8 slides
    create_requests = [
        {"createSlide": {"slideLayoutReference": {"predefinedLayout": "TITLE_AND_BODY"}}}
        for _ in range(8)
    ]

    slides_service.presentations().batchUpdate(
        presentationId=presentation_id,
        body={"requests": create_requests}
    ).execute()

    slides = slides_service.presentations().get(
        presentationId=presentation_id
    ).execute()["slides"]

    # Fill slide text
    fill_requests = []
    for (title, bullets), slide in zip(slides_data, slides):
        title_id = slide["pageElements"][0]["objectId"]
        body_id  = slide["pageElements"][1]["objectId"]
        fill_requests += [
            {"insertText": {"objectId": title_id, "insertionIndex": 0, "text": title}},
            {"insertText": {"objectId": body_id, "insertionIndex": 0, "text": "\n".join("‚Ä¢ "+b for b in bullets)}}
        ]

    slides_service.presentations().batchUpdate(
        presentationId=presentation_id,
        body={"requests": fill_requests}
    ).execute()

    return presentation_id
```

```python
# Run all previous functions with this one
def run_agent():
    topic = input("Enter a topic for your AI-generated presentation: ")

    # Fetch Wikipedia summary
    wiki_text = fetch_wikipedia_summary(topic)
    if "No Wikipedia page" in wiki_text:
        print("‚ö†Ô∏è Could not find a Wikipedia page. Exiting.")
        return

    # Generate slide text using LLM
    slide_text = generate_slide_text(topic, wiki_text)

    # Create Slides from template with generated text
    presentation_title = f"{topic} Slideshow"
    presentation_id = create_google_slides_with_theme(presentation_title, slide_text)

    # Output the link
    print("\nüéâ All done!")
    print(f"üëâ Your presentation is ready:\nhttps://docs.google.com/presentation/d/{presentation_id}/edit")
```

```python
run_agent()
```
